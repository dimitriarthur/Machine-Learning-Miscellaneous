```
from pyspark.ml.feature import StopWordsRemover
from pyspark.ml.feature import NGram
from pyspark.ml.pipeline import Transformer
from pyspark.ml.feature import Tokenizer
from pyspark.ml import Pipeline
from pyspark.sql.functions import col, lower, regexp_replace, split
from pyspark.sql.functions import col, size

class DropRows(Transformer):
    def __init__(self, inputCols):
        self.inputCols = inputCols
    def this():
        this(Identifiable.uid("Dropper_rows"))
    def copy(extra):
        defaultCopy(extra)
    def _transform(self, df):
        return df.na.drop(subset=self.inputCols)
      
class cleanText(Transformer):
    def __init__(self, inputCol, outputCol):
        self.inputCol = inputCol
        self.outputCol = outputCol
    def this():
        this(Identifiable.uid("clean_text"))
    def copy(extra):
        defaultCopy(extra)
    def _transform(self, df):
        def clean_text(c):
          c = lower(c)
          c = regexp_replace(c, "^rt ", "")
          c = regexp_replace(c, "(https?\://)\S+", "")
          c = regexp_replace(c, "[^a-zA-Z0-9\\s]", "")
          return c
        return df.withColumn((self.outputCol), clean_text((self.inputCol)))
stages=[]
ct = cleanText(inputCol='name', outputCol='clean_name')
dr = DropRows(inputCols=['name'])
stages+=[ct, dr]
tokenizer = Tokenizer(inputCol="clean_name", outputCol="vector")
remover = StopWordsRemover()
stopwords = remover.getStopWords() 
remover.setInputCol("vector")
remover.setOutputCol("vector_ns")
ngram = NGram(n=2, inputCol="vector_ns", outputCol="bigrams")
stages+=[tokenizer,remover,ngram]#,st]#]
pipeline = Pipeline(stages=stages)
df_mod = pipeline.fit(df).transform(df)
df_mod = df_mod.where(size(col("vector_ns")) >= 2)
```

# Popular Neighbourhood Groups and Price


Here we are analyzing the neighbourhood groups that, together, represent 97% of the listings (Manhattan, Queens and Brooklyn). 

We can see that: 

* From the boxplots and barplots, Manhattan has the highest mean price, followed by Brooklyn and Queens. 
* Manhattan's mean price is almost the double of Queen's mean price
* There are a lot of outliers, that is, the price vary a lot in each neighbourhood group.

```

df_categorical = (df.select(['neighbourhood_group','price'])).toPandas()
df_categorical = df_categorical[(df_categorical['neighbourhood_group'] == 'Brooklyn')|(df_categorical['neighbourhood_group'] == 'Manhattan')|(df_categorical['neighbourhood_group'] == 'Queens')]
fig, ax = plt.subplots(1, 1, figsize=(10,10))
plt.ylim(0, 1000)
sns.boxplot(y=df_categorical.price, x=df_categorical.neighbourhood_group, ax=ax)
ax.set_title('neighbourhood_group')
```

```
%sql

SELECT neighbourhood_group, mean(price)
FROM AB_NYC_2019_csv
WHERE (neighbourhood_group == 'Brooklyn')or(neighbourhood_group = 'Manhattan')or(neighbourhood_group='Queens')
GROUP BY neighbourhood_group
ORDER BY mean(price)
```

# Room Type and Price

From the plots above we can see:

* Entire home/apt has a consistent number of outliers. It means that the price variance on this one is really strong
* The cheapest type of room is the shared room, followed by private rooms and entire rooms. 
* The mean price for entire rooms/apts is almost 4 times more expensive than shared rooms, and more than 2 times for private rooms.

```
df_categorical = (df.select(['room_type','price'])).toPandas()
df_categorical = df_categorical[(df_categorical['room_type'] == 'Entire home/apt')|(df_categorical['room_type'] == 'Private room')|(df_categorical['room_type'] == 'Shared room')]
fig, ax = plt.subplots(1, 1, figsize=(10,10))
plt.ylim(0, 1000)
sns.boxplot(y=df_categorical.price, x=df_categorical.room_type, ax=ax)
ax.set_title('Room Type}')
```
