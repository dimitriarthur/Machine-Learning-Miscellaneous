```
from pyspark.ml.pipeline import Transformer
from pyspark.sql import functions as f
from pyspark.sql.window import Window
from pyspark.sql.types import FloatType, IntegerType, StringType, DateType
from pyspark.sql.functions import col, avg, radians, dayofweek, dayofmonth, month, hour, year, to_timestamp,weekofyear

class TimestampExtractor(Transformer):
    def __init__(self, inputCol, outputCol1, outputCol2, outputCol3, outputCol4):
        self.inputCol = inputCol 
        self.outputCol1 = outputCol1 
        self.outputCol2 = outputCol2
        self.outputCol3 = outputCol3
        self.outputCol4 = outputCol4
    def this():
        this(Identifiable.uid("TimestampExtractor"))
    def copy(extra):
        defaultCopy(extra)
    def _transform(self, df):
        return df.withColumn(self.outputCol1,dayofweek(f.col(self.inputCol)))\
               .withColumn(self.outputCol2,month(f.col(self.inputCol)))\
               .withColumn(self.outputCol3,year(f.col(self.inputCol)))\
               .withColumn(self.outputCol4,dayofmonth(f.col(self.inputCol)))


class WeekOfYear(Transformer):
    def __init__(self, inputCol,outputCol):
        self.inputCol = inputCol 
        self.outputCol = outputCol
    def this():
        this(Identifiable.uid("WeekOfYear"))
    def copy(extra):
        defaultCopy(extra)
    def _transform(self, df):
        return df.withColumn((self.outputCol), weekofyear(f.col(self.inputCol)))
      
    
class MonthBeginingEnd(Transformer):
  def __init__(self, inputCol,outputCol1,outputCol2):
        self.inputCol = inputCol 
        self.outputCol1 = outputCol1
        self.outputCol2 = outputCol2
  def this():
    this(Identifiable.uid("MonthBegining"))
  def copy(extra):
    defaultCopy(extra)
  def _transform(self, df):
    return df.withColumn(self.outputCol1, f.when(f.col(self.inputCol) <= 7,1).otherwise(0))\
             .withColumn(self.outputCol2, f.when(f.col(self.inputCol) >=24, 1).otherwise(0))
  
class YearQuarter(Transformer):
  def __init__(self, inputCol,outputCol):
        self.inputCol = inputCol 
        self.outputCol = outputCol
  def this():
    this(Identifiable.uid("yrqrtr"))
  def copy(extra):
    defaultCopy(extra)
  def _transform(self, df):
    return df.withColumn(self.outputCol, f.when(f.col(self.inputCol) <=3,1)\
             .otherwise(f.when(f.col(self.inputCol) <=6,2)\
             .otherwise(f.when(f.col(self.inputCol) <=9,3)\
             .otherwise(4))))
  
  
class WeekendExtractor(Transformer):
  def __init__(self, inputCol,outputCol):
        self.inputCol = inputCol 
        self.outputCol = outputCol
  def this():
    this(Identifiable.uid("wkndxt"))
  def copy(extra):
    defaultCopy(extra)
  def _transform(self, df):
    return df.withColumn(self.outputCol, f.when((f.col(self.inputCol) == 6)|(f.col(self.inputCol)==7),1).otherwise(0))
```

```
from pyspark.ml import Pipeline
tse = TimestampExtractor(inputCol='date', outputCol1='dayofweek', outputCol2='month', outputCol3='year', outputCol4='dayOfMonth')
woy = WeekOfYear(inputCol='date',outputCol='weekOfYear')
mb = MonthBeginingEnd(inputCol='dayOfMonth', outputCol1='monthBegining', outputCol2='monthEnd')
yq = YearQuarter(inputCol='month', outputCol='YearQuarter')
we = WeekendExtractor(inputCol='dayofweek',outputCol='weekend')
stages = [tse,mb,yq,woy,we]
pipeline = Pipeline().setStages(stages)
pipelineModel = pipeline.fit(df)
df_mod = pipelineModel.transform(df)
display(df_mod)
```

### Seasonal Decomposition

This is a statistical approach to analyze the time series in multiple components. We can see that the series is probably not stationary due to its increasing trend. We can also see that the July/November peak pattern is also visible.

```
import pandas as pd
df_to_decompose = df_mod.select("date","sales","year").toPandas()
df_to_decompose['date'] = pd.to_datetime(df_to_decompose.date,format="%Y.%m.%d")
df_to_decompose = df_to_decompose[["date", "sales","year"]].set_index("date")
df_to_decompose = df_to_decompose.resample("M").sum()
```

```
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df_to_decompose.groupby('date').sales.sum(), model='additive')
fig, axes = plt.subplots(ncols = 1, nrows = 4, sharex = False, figsize = (11,10))
result.observed.plot(ax = axes[0], legend = True)
plt.legend()
axes[0].set_title('Monthly Decomposition',fontsize=14,fontweight='bold')
axes[0].set_ylabel('Original',fontsize=14,fontweight='bold')

result.trend.plot(ax = axes[1], legend = True)
axes[1].set_ylabel('Trend',fontsize=14,fontweight='bold')

result.seasonal.plot(ax = axes[2], legend = True)
axes[2].set_ylabel('Seasonal',fontsize=14,fontweight='bold')

result.resid.plot(ax = axes[3], legend = True)
axes[3].set_ylabel('Residual',fontsize=14,fontweight='bold')
```

### Dickey-Fuller Test

Based on the p-value, we can infer that the time series is non-stationary. To deal with this in the modeling, we should use a model that deals with this by diferencing the series, such as the ARIMA or SARIMAX model.

```
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries, window = 12, cutoff = 0.01):

    #Determing rolling statistics
    rolmean = timeseries.rolling(window).mean()
    rolstd = timeseries.rolling(window).std()
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC', maxlag = 20 )
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    pvalue = dftest[1]
    if pvalue < cutoff:
        print('p-value = %.4f. The series is likely stationary.' % pvalue)
    else:
        print('p-value = %.4f. The series is likely non-stationary.' % pvalue)
    
    print(dfoutput)
  
test_stationarity(df_mod_2.y)
```

### Modeling - ARIMA

We are modeling the time series related to the store 2, item 15. 

* The series is not stationary, so we are setting d = 1
* To check the AR model oder, p, we pay attention to the PACF plot. As one can see, after the 7th lag, there is some oscillation, but there aren't significant autocorrelation values above the set threshold. So p = 7
* To check the number of moving average terms, we look at the ACF plot.

```
df_mod_2['date'] = pd.to_datetime(df_mod_2.date,format="%Y.%m.%d")
df_mod_2 = df_mod_2[["date", "sales"]].set_index("date")
df_mod_2.index.rename("ds", inplace=True)
df_mod_2.rename(columns={'sales':'y'}, inplace=True)
```

```
from pandas import datetime
from matplotlib import pyplot
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
 

X = df_mod_2.values
size = int(len(X) * 0.95)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(7,1,3))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('prediction=%f, test=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('MSE: ', error)
```

### Validation Data Results

The ARIMA(7,1,3) model seems to be relatively good at making predictions on the test data. We have separated 90 days to the validation set, and the rest to train the model. It seems to have captured most of the trends on the data, although the MSE is still high.

```
plt.figure(figsize=(20,8))
plt.title('Validation - ARIMA Model', fontweight='bold', fontsize=14)
plt.xlabel('Month',fontsize=14,fontweight='bold')
plt.ylabel('Amount',fontsize=14,fontweight='bold')
plt.plot(test, label='test data')
plt.plot(df_mod_2.index[size:len(X)].to_native_types().tolist(),predictions, color='red', label='predictions')
plt.legend()
plt.show()
```
