# Pipeline - Feature Engineering

#### Custom Classes:
* Drop Columns: take specific columns of the dataset
* MeanImputation: Fill the age feature missing values with the average value for the Pclass and Sex.
* AgeRange: categorically classify the age
* FamilySize: returns the sum of SibSp and Parch
* FamilyClass: categorically classify the family size

```
from pyspark.ml.pipeline import Transformer
from pyspark.sql import functions as f
from pyspark.sql.window import Window
from pyspark.sql.functions import col, avg


class DropColumns(Transformer):
    def __init__(self, outputCol):
        self.outputCol = outputCol 
    def this():
        this(Identifiable.uid("Dropper"))
    def copy(extra):
        defaultCopy(extra)
    def _transform(self, df):
        return df.select(self.outputCol)

class MeanImputation(Transformer):
    # based on the Pclass and Sex, the mean is imputed to the missing values
    def __init__(self, inputCol, outputCol):
        self.inputCol = inputCol 
        self.outputCol = outputCol 
    def this():
        this(Identifiable.uid("imputer"))
    def copy(extra):
        defaultCopy(extra)
    def _transform(self, df):
        w = Window().partitionBy(self.inputCol)
        return df.withColumn(self.outputCol,when(col(self.outputCol).isNull(),avg(col(self.outputCol)).over(w)).otherwise(col(self.outputCol)))

class AgeRange(Transformer):
    def __init__(self, inputCol, outputCol):
        self.inputCol = inputCol 
        self.outputCol = outputCol 
    def this():
        this(Identifiable.uid("ageRange"))
    def copy(extra):
        defaultCopy(extra)
    def _transform(self, df):
        df = df.withColumn(self.outputCol, f.when(f.col(self.inputCol) < 10, 'kid'))
        df = df.withColumn(self.outputCol, f.when(((f.col(self.inputCol) >= 10)&(f.col('Age') < 20)), 'teen').otherwise(f.col(self.outputCol)))
        df = df.withColumn(self.outputCol, f.when(((f.col(self.inputCol) >= 20)&(f.col('Age') < 40)), 'adult').otherwise(f.col(self.outputCol)))
        df = df.withColumn(self.outputCol, f.when((f.col(self.inputCol) >= 40), 'elder').otherwise(f.col(self.outputCol)))
        return df
      
class FamilySize(Transformer):
    def __init__(self, inputCol1, inputCol2, outputCol):
        self.inputCol1 = inputCol1 
        self.inputCol2 = inputCol2 
        self.outputCol = outputCol 
    def this():
        this(Identifiable.uid("family_size"))
    def copy(extra):
        defaultCopy(extra)
    def _transform(self, df):
        df = df.withColumn(self.outputCol, col(self.inputCol1) + col(self.inputCol2) + 1)
        return df            
      

class FamilyClass(Transformer):
    def __init__(self, inputCol, outputCol):
        self.inputCol = inputCol 
        self.outputCol = outputCol 
    def this():
        this(Identifiable.uid("family_class"))
    def copy(extra):
        defaultCopy(extra)
    def _transform(self, df):
        df = df.withColumn(self.outputCol, f.when(f.col(self.inputCol) < 2, 'alone'))
        df = df.withColumn(self.outputCol, f.when((f.col(self.inputCol) >=2)&(f.col(self.inputCol)<5), 'small').otherwise(f.col(self.outputCol)))
        df = df.withColumn(self.outputCol, f.when((f.col(self.inputCol) >=5)&(f.col(self.inputCol)<7), 'medium').otherwise(f.col(self.outputCol)))
        df = df.withColumn(self.outputCol, f.when((f.col(self.inputCol) >=7), 'big').otherwise(f.col(self.outputCol)))
        return df      
           
```

```
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, VectorAssembler, OneHotEncoder
from pyspark.ml.classification import RandomForestClassifier

stages = []

dc = DropColumns(outputCol = ["Survived","Pclass","Sex","Age","SibSp","Parch","Fare"])
mi = MeanImputation(inputCol=["Pclass","Sex"], outputCol="Age")
ar = AgeRange(inputCol='Age', outputCol='Age_Range')
fs = FamilySize(inputCol1='Parch',inputCol2='SibSp',outputCol='family_size')
fc = FamilyClass(inputCol='family_size',outputCol='family_class')
stages += [dc,mi,ar,fs,fc]
stage_cols = ['Sex','family_class','Pclass','Age_Range']


for column in stage_cols:
  stringIndexer = StringIndexer(inputCol = column, outputCol = column + "idx")
  encoder = OneHotEncoder(inputCols = [stringIndexer.getOutputCol()], outputCols = [column + "catVec"])
  stages+=[stringIndexer,encoder]

feature = VectorAssembler(inputCols=["Age","SibSp","Parch","SexcatVec","family_classcatVec","PclasscatVec","Age_RangecatVec"],outputCol="fts")
stages+=[feature]
```

```
pipeline = Pipeline(stages=stages)
df = pipeline.fit(df).transform(df)
```

```
(trainingData, testData) = df.randomSplit([0.8, 0.2],seed = 11)
```

# Modeling

#### Random Forest

This is the best model found out. After **numTress** > 100, there is not much of improvement on the accuracy score and the algorithm takes much longer time to run. The **maxDepth** value seemed ok being set as 4. As if it represents the depth of each tree and hence the increasing number of splits, when it is too large, the model tends to overfit. **maxBins** allows the algorithm to consider more split candidates, but it needs to be at least the number of  maximum number of categories M for any categorical feature. The value chosen was 50.

```
rf = RandomForestClassifier(labelCol = 'Survived', 
                            featuresCol = 'fts', 
                            numTrees = 100, 
                            maxDepth = 4, 
                            maxBins = 50)
rf_model = rf.fit(trainingData)
```

#### Gradient Boost Tree
The parameters chosen after some experimentation follow the same logic of the RF. **maxIter** didn't seem to improve much the results during the tests.

```
from pyspark.ml.classification import GBTClassifier
gbt = GBTClassifier(labelCol="Survived", featuresCol="fts",maxIter=10,maxDepth=5)
gbt_model = gbt.fit(trainingData)
```

```
rf_prediction = rf_model.transform(testData)
gbt_prediction = gbt_model.transform(testData)
```

```
def plot_confusion_matrix(cm,
                          target_names,
                          title='Confusion matrix',
                          cmap=None,
                          normalize=True):

    import matplotlib.pyplot as plt
    import numpy as np
    import itertools

    accuracy = np.trace(cm) / float(np.sum(cm))
    misclass = 1 - accuracy

    if cmap is None:
        cmap = plt.get_cmap('Blues')

    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title,fontweight='bold',fontsize=14)
    plt.colorbar()

    if target_names is not None:
        tick_marks = np.arange(len(target_names))
        plt.xticks(tick_marks, target_names, rotation=0)
        plt.yticks(tick_marks, target_names)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]


    thresh = cm.max() / 1.5 if normalize else cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if normalize:
            plt.text(j, i, "{:0.4f}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
        else:
            plt.text(j, i, "{:,}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")


    plt.tight_layout()
    plt.ylabel('True label',fontweight='bold',fontsize=14)
    plt.xlabel('Predicted label\naccuracy={:0.4f}; misclass={:0.4f}'.format(accuracy, misclass),fontweight='bold',fontsize=14)
    plt.show()
    
```

#### Model's Analysis
About the models, we saw that GBT performed slightly better than RF. This can be explained by:
* Random forests are a large number of trees, combined (using averages or "majority rules") at the end of the process.
* Gradient boosting machines also combine decision trees, but start the combining process at the beginning, instead of at the end.
* Random forests builds each tree independently while gradient boosting builds one tree at a time. This additive model (ensemble) works in a forward stage-wise manner, introducing a weak learner to improve the shortcomings of existing weak learners. 
* Combining results: random forests combine results at the end of the process (by averaging or "majority rules") while gradient boosting combines results along the way.
* Gradient boosting, after we tuned up some few parameters, performed better than the RF. The good performance can also be explained by the low noise on the dataset, which could be a big problem for GBT.
