# Modeling

Here we are using only a Random Forest Model in order to explore properly the two parameters: maxDepth and numTrees. From the results, we can see that the optimal value to train this ensemble method is maxDepth=15 and numTrees=20. This provides a reduction on the variance of the model, avoiding it to overspecialize on the training data.

```
(trainingData, testData) = df_mod_2.randomSplit([0.8, 0.2],seed = 11)
```

```
trainingData.count()
```

```
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator

rmses = [[],[],[]]
depths = []
pos = 0
for numTree in range(10,25,5):
  for maxDep in range(5,35,5):
    rf = RandomForestRegressor(featuresCol='fts',labelCol='SalePrice', maxDepth=maxDep, numTrees=numTree)
    rf_model = rf.fit(trainingData)
    pred_rf = rf_model.transform(testData)
    evaluator = RegressionEvaluator(metricName="rmse", labelCol="SalePrice", predictionCol=rf.getPredictionCol())
    rf_rmse = evaluator.evaluate(pred_rf)
    rmses[pos].append(rf_rmse)
    depths.append(maxDep)
  pos +=1
```

```

plt.plot(depths[0:6],rmses[0])
plt.plot(depths[0:6],rmses[1])
plt.plot(depths[0:6],rmses[2])
plt.xlabel('Depth', fontsize = 13, weight = 'bold')
plt.ylabel('RMSE', fontsize = 13, weight = 'bold')
plt.legend(['numTress = 10', 'numTress = 15', 'numTress = 20'], loc='upper right')
plt.show()
```

# Conclusions

#### About the data
* Although all the features are important on the real state market, most of them have not been useful to apply linear regression on the target variable. That is why we used a diminished number of features.
* There's a lot of outliers on the main features. This can lead the model to inconclusive decisions.
* Most of the houses have a medium overall quality in a range of 5 to 7 (from 0 to 10).
* Mos of the houses were built close to the 2000's. This is also true to the year of remodeling. 
* Houses on streets that are paved have the most number of outliers in respect to sale price. 
* Buildings on the NoRidge and NrdgHt have the highest mean prices and also the greates variance on the sale price. 
* OverallQual is a preponderant factor on the sale price increase.

#### About the model
* We have satisfactorily reached a hyperparameter tuning that lead to a great reduction on the RMSE. This is due to the number of alternative trees used on the ensemble process, softening specific models biases and also diminishing the variance. Also a Depth of 15 proved to be a good value, taking into consideration the computational complexity and the overfitting on the model. Also the feature selection seemed relevant. From more than 80 features, we could use less than 20 which were really correlated to the target. The results in terms of RMSE were really good, and this can be explained by the gain of information at each split, once again due to the good feature selection that was applied. 

#### For Future Improvement
* We have dealt with a good portion of the outliers. Althoughm, a lot of them remained. Taking a more agressive removel approach might help even more the model.
* Due to time restraints, cross-validation could not be implemented. Appplying this technique could make the model experience even more variance on the dataset than simply splitting the data on 80/20%.
* As one can see on the EDA, the SalePrice feature is considerably negatively skewed. Some log transform could be applied to it, which would help the linear regression process.
